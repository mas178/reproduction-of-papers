```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file_name = '../../data/degrees.csv'
df = pd.read_csv(csv_file_name, header=None, names=['hop_learning', 'degree'])
```

```python
for hop_learning in range(1, 7):
    if hop_learning < 4:
        p = sns.histplot(df[df.hop_learning == hop_learning].degree)
    else:
        p = sns.histplot(df[df.hop_learning == hop_learning].degree, binrange=(2, 999), binwidth=50)
    p.set_title('hop_learning = {}'.format(hop_learning))
    plt.show()
```
