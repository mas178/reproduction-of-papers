```python
import pandas as pd
import seaborn as sns
```

```python
def show_graph(csv: str) -> None:
    df = pd.read_csv(csv)
    g = sns.relplot(kind='line', x='b', y='cooperation_rate', hue='network_type', col='update_rule', data=df)
    g.set_axis_labels('b/c', 'Cooperator Rate')
    g.set(ylim=(0, 1))
    display(g)
```

```python
show_graph('FUM.csv')
```
