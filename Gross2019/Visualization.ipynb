```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def load_data(csv_file_name: str) -> pd.DataFrame:
    return pd.read_csv(
        csv_file_name,
        header=None,
        names=['Generation', 'Heider', 'Friend_Focused', 'All_D'],
        dtype={0: int, 1: int, 2: int, 3: int},
        index_col=0
    )

def aggregate_df(df: pd.DataFrame) -> pd.DataFrame:
    df["KG"] = df.index / 1000 + 1
    df["KG"] = df["KG"].round().astype(int)
    return df.groupby(['KG']).mean()
```

```python
%%time
df1 = aggregate_df(load_data("data/20220318_151944.csv"))
df2 = aggregate_df(load_data("data/20220318_152808.csv"))
df3 = aggregate_df(load_data("data/20220318_160421.csv"))
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(12.0, 4.0))
sns.lineplot(data=df1, ax=axes[0])
sns.lineplot(data=df1.iloc[300:350], ax=axes[1])

fig, axes = plt.subplots(ncols=2, figsize=(12.0, 4.0))
sns.lineplot(data=df2, ax=axes[0])
sns.lineplot(data=df2.iloc[300:350], ax=axes[1])

fig, axes = plt.subplots(ncols=2, figsize=(12.0, 4.0))
sns.lineplot(data=df3, ax=axes[0])
sns.lineplot(data=df3.iloc[300:350], ax=axes[1])
```
