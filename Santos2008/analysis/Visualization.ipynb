# Visualization

```julia
using CSV
using DataFrames
using Plots
```

```julia
df = CSV.read(
    "../data/output_20210924_002221.csv",
    DataFrame;
    header=[:network_type, :cost_model, :η, :c_rate],
    types=[Symbol, Symbol, Float64, Float64]
)[:, [:cost_model, :η, :c_rate]]
```

```julia
df_fixed_cost = df[df.cost_model .== :fixed_cost, :]
df_variable_cost = df[df.cost_model .== :variable_cost, :]

plot(
    df_fixed_cost.η,
    df_fixed_cost.c_rate;
    label = "Fixed cost",
    xlabel = "η",
    ylabel = "Fraction of cooperators",
    xlims = [0.5, 0.8],
    ylims = [0.0, 0.8],
    title = "Evolution of cooperation in networked PGGs",
    fmt = :png,
    legend = :bottomright
)
plot!(df_variable_cost.η, df_variable_cost.c_rate; label = "Variable cost")
```
