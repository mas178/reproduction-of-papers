```python
import pandas as pd
import seaborn as sns
```

```python
def draw(csv_file_name: str) -> None:
    df = pd.read_csv(csv_file_name, header=None, names=['m', 'n', 'b/c', 'cooperation_rate'])

    fig_n10 = sns.relplot(
        data=df[df.n == 10],
        x='b/c',
        y='cooperation_rate',
        col='m',
        kind='line',
        legend=False
    )

    fig_m10 = sns.relplot(
        data=df[df.m == 10],
        x='b/c',
        y='cooperation_rate',
        col='n',
        kind='line',
        legend=False
    )

    fig_n10.fig.subplots_adjust(top=.9)
    fig_n10.fig.suptitle('n = 10')

    fig_m10.fig.subplots_adjust(top=.9)
    fig_m10.fig.suptitle('m = 10')
```

```python
# w = 0.1, q = 0.001
draw('data/20220224_112507.csv')
```

```python
# w = 0.01, q = 0.01
draw('data/20220223_142641.csv')
```

```python
# w = 0.1, q = 0.1
draw('data/20220223_223515.csv')
```
